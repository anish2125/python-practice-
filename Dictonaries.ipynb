### Introduction to dictonaries
    Dictonaries are the unordered collection of items. They store data in  key value pairs. Key must be unique and immutable(e.g. strings, numbers and tuples) while the data can be of anytype
    

```python
## creating dictonariwa
empty_dict={}
print(type(empty_dict))
```

```python
empty_dict= dict()
empty_dict
```

```python
student = {"name":"Anish" , "age":22 , "grade":22}
print(student)

print(type(student))
```

```python
## Single key is always used insted of multiple keys


student = {"name":"Anish" , "age":22 , "name":22}
print(student)
```

```python

## Accessing dictonaries elenent

student = {"name":"Anish" , "age":22 , "grade":'A'}

print(student['grade'])

print(student['age'])

print(student["name"])

## Accessing usig get() method provided by the dictonaries


print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name', "is not available"))
```

```python
## Iterating 0ver dictonaries
## You can use loops to iterate And we can iterate over dictonaries, keys, values or items

## iteratng over kry

for  keys in student.keys():
    print(keys)
```

```python
## Nested Disctonaries
stdents={
    "student1":{"name":"Anish","age":32},
    "studen2":{"name":"Pallavi", "age":36}
}

print(stdents)
```

```python
## Access nested disctonaries element

print(stdents["studen2"]["name"])
print(stdents["student1" ]["age"])
```

```python
## Iterating over nested disctonariefor 

for student_id, student_info in stdents.items():
    print(f"{student_id}:{student_info}")

    for  key, value in student_info.items():
        print(f"{key}:{value}")
        
```

```python
## Dictionary Comphrension
squares = {x : x**2 for x in range(5)}
print(squares)
```

```python
### conditional dictionaries comphrension

even = {x : x**2 for x in range(10) if x %2==0}

print(even)
```

```python
## Pratical exxmaples

## Use a dictionary to count the freqncy  of element  in a list

numbers = [1,2,2,3,3,3,4,4,4,4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## Merge two dictonaries into one

dict1={"a":1, "b":2}
dict2={"b":3, "c":4}

merged_dict ={**dict1, **dict2}
print(merged_dict)
```
