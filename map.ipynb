### The map() Function in Python
The map() function applies a given function to all items in an input lust(orany other iterable) and  returns a map object. This is particularlyb used fot transforming a data in a list comprehensively

```python
def sqaure(x):
    return x**2
sqaure(5)
```

```python
number=[1,2,3,4,5,6,7,8,9]
result= list(map(lambda x : x**2,number))
print(list(result))
```

```python
number= [1,2,3,4,5]
result = list(map(lambda x : x**3,number))
print(list(result))
```

```python
###  can we map with multiple iterables? yes we can
number1 =[1,2,3,4,5]
number2 =[6,7,8,9,10]
result= map(lambda x,y: x*y, number1,number2)
print(list(result))
```

```python
a = [12,44,56,67,78]
b = [45,56,78,89,9]
result = map(lambda x,y:x-y, a,b)
print(list(result))
```

```python

## map() to convert list of sting to integers
str_number = ['1','2','3','4','5']
int_number = list(map(int,str_number))
print(int_number)
```

```python
str_numbers = ['10','20','30','40','50']
int_number= list(map(int,str_numbers))
print(int_number)
```

```python
word = ['python','java','c++','ruby']
upper_word= list(map(lambda x :x.upper(),word))
print(upper_word)
```
