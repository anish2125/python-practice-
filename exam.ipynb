## temperature conversion

```python
def convert_temperature(temp,unit):
    """Convert temperature between Celsius and Fahrenheit."""
    if unit == "C":
        return temp * 9/5 + 32  ## celsius to fahrenheit
    elif unit == "F":
        return (temp - 32) * 5/9  ## fahrenheit to celsius
    else:
        return None

print(convert_temperature(25, 'C'))  ## Output: 77.0
print(convert_temperature(77, 'F'))  ## Output: 25.0
```

### Password strength checker

```python
def is_strong_password(password):
    """Check if the password is strong."""
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in "!@#$%^&*()-_=+[]{}|;:'\",.<>?/`~" for char in password):
        return False
    return True

## Calling the function
print(is_strong_password("Anish@123!"))
print(is_strong_password("weakpass"))
    
```

### Calculate the Total cost of items in the shopping bag

```python
def calculate_total_cost(cart):
    total_cost = 0 
    for item in cart:
        total_cost += item['price'] * item['quantity']
    return total_cost
## example cart data
cart =[
    {'name':'Apple', 'price':1.2, 'quantity':3},
    {'name':'Banana', 'price':0.5, 'quantity':6},
    {'name':'Orange', 'price':0.8, 'quantity':4}
]
calculate_total_cost(cart)
```

### Program  to check palindrome number

```python
def is_palindrome(string):
    """Check if the given string is a palindrome or not"""
    reversed_string = string[::-1]
    ## check if original string is same as the reversed string or not
    if string == reversed_string:
        return True
    else:
        return False

# calling the function
word = input("Enter a string: ")
if is_palindrome(word):
    print(f"{word} is a palindrome")
else:
    print(f"{word} is not a palindrome")
```

### Factorial of a number using recursion

```python
def factorial(n):
    # base case
    if n == 0 or n == 1:
        return 1
    else:
        # recursive case
        return n * factorial(n - 1)


# take input from user
num = int(input("Enter a number: "))

# function call
result = factorial(num)

print("Factorial of", num, "is:", result)
```
