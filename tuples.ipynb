```python
## Packing and unpacking tuple
##packing
packed_tuple= 1,"Hello",3.14
print(packed_tuple)
```

```python
## unpacking a tuple

a,b,c=packed_tuple
print(a)
print(b)
print(c)
```

```python
## unpacking with *

number =(1,2,3,4,5,6)
first,*middle,last = number
print(first)
print(middle)
print(last)
```

```python
## Nested tuple
lst= [[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,True]]  ##nested list
lst[0][3]
lst[2][0:3]
```

```python
lst= [[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,True)]  ##nested list
lst[0][3]
lst[2][0:3]
```

```python
##nested tuple
nest_tuple=((1,2,3),("a","b","c"),(True, False))

## access the elements inside a tuple
print(nest_tuple[0])
print(nest_tuple[1][2])
```

```python
## Iterating under nested tuples

for sub_tuple in nest_tuple:
    for item in sub_tuple:
        print(item , end =" ")
    print()
```
