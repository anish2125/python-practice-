```python
##Nested condtional statements - 
## You can place one  or more if, elif, or else statements inside another if, elif, or else statement to create nested conditional statements.
 
 ## number to check even or negatue

num = int(input("Enter the number"))
if(num>0):
    print("Number is positive")
    if num%2==0:
        print("Number is even")
    else: 
        print("Number is odd")
else:
    print("The number is either 0 or negative")
```

```python
## Pratical Examples
## Determine if a year is a leap year using nested conditioy
year = int(input("Enter the year"))

if year % 4 == 0:
    print(year,"Year is leap year")
    if year % 400 == 0:
        print(year,"Year is leap year")
        if year % 100 ==0:
            print(year,"Year is a leap year")
else:
    print(year,"Year is not leap year")
```

```python
## simple calculator using nested if else
num1 = float(input("Enter the first Number:"))
num2 = float(input("Enter the second Number:"))

operation = input("Enter operation (+,-,*,/)")
if operation == '+':
    result = num1 + num2
elif operation == '-':
    result = num1 - num2
elif operation == '*':
    result = num1 * num2
elif operation == '/':
    if num2 != 0:
        result = num1/num2
    else:
        result = "Error! Division by zero."
else:
    result ="Invalid Operation."

print(result)
```
