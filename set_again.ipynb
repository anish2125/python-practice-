```python
## mathematical operation
set1 = {1,2,3,4,5,6}

set2= {4,5,6,7,8,9}

union_set= set1.union(set2)
print(union_set)

interecr_set = set1.intersection(set2)
print(interecr_set)

set1.intersection_update(set2) ## this mean that the common elemen6t from set1 and set2 we get and that all the elements are updated in set1
print(set1)
```

```python
set1 = {1,2,3,4,5,6}

set2= {4,5,6,7,8,9}

print(set1.difference(set2))
```

```python
set2.difference(set1) ## common elements are being removed
```

```python
## sets  method
set1={1,2,3}
set2={3,4,5}

print(set1.issubset(set2))  ##  this means all that element are available in set1 is it also availabelle in set 2

print(set2.issubset(set1))

print(set1.issuperset(set2))
```
