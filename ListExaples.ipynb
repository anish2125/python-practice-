```python
# Simple To-Do List Manager

todo_list = []   # each item will be a dictionary: {"task": "Buy milk", "done": False}

def show_menu():
    print("\n--- TO-DO LIST MENU ---")
    print("1. View tasks")
    print("2. Add task")
    print("3. Mark task as completed")
    print("4. Delete task")
    print("5. Exit")

def view_tasks():
    if not todo_list:
        print("\nNo tasks in the list.")
        return
    
    print("\nYour tasks:")
    for index, item in enumerate(todo_list, start=1):
        status = "✅ Done" if item["done"] else "❌ Not done"
        print(f"{index}. {item['task']} [{status}]")

def add_task():
    task = input("\nEnter the task: ")
    todo_list.append({"task": task, "done": False})
    print("Task added!")

def mark_completed():
    view_tasks()
    if not todo_list:
        return
    
    try:
        choice = int(input("\nEnter task number to mark as completed: "))
        if 1 <= choice <= len(todo_list):
            todo_list[choice - 1]["done"] = True
            print("Task marked as completed!")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Please enter a valid number.")

def delete_task():
    view_tasks()
    if not todo_list:
        return
    
    try:
        choice = int(input("\nEnter task number to delete: "))
        if 1 <= choice <= len(todo_list):
            removed = todo_list.pop(choice - 1)
            print(f"Deleted task: {removed['task']}")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Please enter a valid number.")

# Main loop
while True:
    show_menu()
    option = input("\nChoose an option (1-5): ")
    
    if option == "1":
        view_tasks()
    elif option == "2":
        add_task()
    elif option == "3":
        mark_completed()
    elif option == "4":
        delete_task()
    elif option == "5":
        print("Goodbye!")
        break
    else:
        print("Invalid choice, please try again.")
```
