### Lambda Function in Python
Lambda functions are small anonymous functions defined using lambda keyword. They can have any number of arguments but only one expression. They are commonly used for short operations or as arguments to higher arguments to higher-order functions

```python
lambda arguments: expression
```

```python
def addition(a,b):
    return a+b
```

```python
addition(5,7)
```

```python
additiion = lambda a,b :a+b ## This isn the way to define the lambda function
type(additiion)
additiion(5,7) ## here it is calling and passing the values and printing
```

```python
def even(num):
    if num%2 ==0:
        return True
even(4)
```

```python
## hereits is showing the above even function by using the lambda function
even1= lambda num:num%2==0
even1(41)
```

```python
def addition(x,y,z):
    return x+y+z
addition(5,6,7)
```

```python
addtiton1= lambda x,y,z : x+y+z
addtiton1(5,6,7)
```

```python
## map()   function with lambda - it applies a function to all the items in an imput lists
numbers= [12,33,42,3,4,51,6,7,8,9]
square = map(lambda x : x**2,numbers)
print(list(square))
```

```python
## map with multiple iterables
## we can use mmap() with multiple iterables if the function we are using takes more than one arguments

number1 =[1,2,3,4,5]
number2 =[6,7,8,9,10]
result= map(lambda x,y : x+y, number1,number2)
print(list(result))
```
