```python
## Packing and unpaCKING tuple

packed_tuple= 1,"Hello",3.14
print(packed_tuple)
```

```python
## Unpacking tupke

a,b,c = packed_tuple
print(a)
print(b)
print(c)
```

```python
## Unpacking with *

numbers= (1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(*middle)
print(last)
```

```python
## nessted Tuple
lst=[[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,"c"]]
print(lst)

print([lst[0]])
print([lst[0][2]])
print([lst[2][0:3]])
```

```python
nested_tuple=((1,2,3),("a","b","d"),(True,False))

##accesss the element inside the tuple

print(nested_tuple[0])

print(nested_tuple[2])

print(nested_tuple[1][2])
```

```python
## iterating over the nested tuple

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end=" ")
        print()
```
