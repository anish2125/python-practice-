### Sets 

Sets are a built-in data type in Python used to store collection of unique items. They are unordered, meaning that the element do not follow the specific order, and they do not allow the duplicate element. Sets are useful for membership tests, eleminating duplicate entries, and performing mathematical operations like union, intersection, difference and symmertric difference

```python
## create a set
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set= set()
print(type(my_empty_set))
```

```python
my_set= set({1,2,3,4,5,6})
print(my_set)
```

```python
my_empty_set= set({1,2,3,4,5,6,5,6})
print(my_empty_set)
```

```python
##  Basics  set operations
## Adding and Removing Elements

my_set.add(7)
print(my_set)
```

```python
my_set.remove(7)
print(my_set)
```

```python
my_set.remove(19)
```

```python
my_set.discard(11)

print(my_set)
```

```python
#3 pop methods

removes_element= my_set.pop()
print(removes_element)
print(my_set)
```

```python
## clear all emenment

my_set.clear()
print(my_set)
```

```python
## set membership test

## It checks weather that element is present in that set or not

my_set = {1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## Mathematical operations

set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

### Union Operation

union_set = set1.union(set2)
print(union_set)
```

```python
set3 = {44,87,78,22}
set4 = {65,51,77,99}
union = set3.union(set4)
print(union)
```

```python
## Mathematical operations

set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

### Union Operation

union_set = set1.union(set2)
print(union_set)

## Intersection


Intersection = set1.intersection(set2)
print(Intersection)

set1.intersection_update(set2)
print(set1)
## Intersection_update will update  all the value of operation and will update in set1

```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

## Difference

print(set1.difference(set2))
```

```python
## sets metbod
set1={1,2,3,4,5,}
set2 = {3,4,5}

## is subset


print(set1.issubset(set2))
print(set1.issuperset(set2))
```

```python
## Counting unique words in text

text = "In this tutorial we are discussing about sets"

words= text.split()
print(words)

## convert list of words to set to get unique words

unique_words= set(words)
print(unique_words)
print(len(unique_words))
```
