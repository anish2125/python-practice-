```python
## while loop
 ## The while loop continues to executes as long as the condition ia true.
  
count = 0
while count<5:
    print(count)
    count = count+1 
```

```python
count = 2
while count%2==0:
    print(count)
    count = count+1
```

```python
#Loop Control Statements

## break
## The break statement exits the loop  prematurely

for i in range(10):
    if i ==5:
        break
    print(i)
```

```python
## continue
## The continue statements  skips the currnt iteration and continues with the next
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## passt
## The pass statement is null operation; it does nothing

for i in range(5):
    if i == 3:
        pass
    print(i)
```

```python
## Nested loops
## a loop inside loop

for i in range(3):
    for j in range(2):
         print(f"i:{i} and j:{j}")
```

```python
for i in range(5):
    for j in range(4):
        print(f"i:{i} and j:{j}")
```

```python
## Calculates the sum of first N natural numbers using a while and for loop\

## while loop

n = 10
sum =0
count =1

while count<=n:
    sum = sum + count
    count = count+ 1
print("Sum of first 10 natural number:", sum)
```
