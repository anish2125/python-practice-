## Functions


 A function is a block of code that perfoms specifc task. Functions help in prganizing code, reusing code, and improving readability

```python
## syntax

def function_name(parameters):
    """Docstring explaining the function."""
    ## function body
    return expression
```

```python
## why fn requiew
num = 24
if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

```python
def even_or_odd(number):
    """Determine if a number is even or odd."""
    if number % 2 == 0:
     print(f"{number} is even")
    else:
     print(f"{number} is odd")
```

```python
## calling fn
even_or_odd(10)
even_or_odd(7)  
even_or_odd(0)
even_or_odd(-4) 
    
```

```python
## function with multiple parameters
def add_numbers(a, b):
    """Return the sum of two numbers."""
    return a + b    
```

```python
add_numbers(5, 10)
```

```python
## default parameter values
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}")
greet("Anish, Good Morning")
```

```python
### Variable length parameters
### Positional and Keyword Arguments

## for variable length parameters we often use  *args and **kwargs but you can use any name
def print_numbers(*anish):
    for num in anish:
        print(num)
```

```python
print_numbers(1, 2, 3, 4, 5,"anish")
```

```python
## positional arguments
def print_numbers(*args):
    for num in args:
        print(num)
```

```python
print_numbers(1, 2, 3, 4, 5,"anish")
```

```python
### keyword arguments

def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
     
```

```python
print_details(name="Anish", age="25", city="New York")
```

```python
def print_details(*args, **kwargs):
    for val in args:
        print(f "positional arg: {val}")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_details(1,2,3,4, name="Anish", age="25", city="New York")
```
